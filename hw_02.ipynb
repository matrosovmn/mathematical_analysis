### 1. Как соотносятся понятия “множество” и “последовательность”? (в ответе использовать слова типа: часть, целое, общее, частное, родитель, дочерний субъект и т. д.)



Множество состоит из элементов объединенных общим свойством.
* порядок записи элементов множества не влияет на само множество {6,11}={11,6}
* для множества не определено число вхождений одинаковых элементов
* множество может быть пустым и непустым, упорядоченным и неупорядоченным, конечным и бесконечным,

Последовательность это частный случай множества, в котором набор объектов пронумерован, допускаются повторения объектов и порядок объектов имеет значение. 

### 2. Прочитать высказывания математической логики, построить их отрицания и установить истинность.

**1. ∀y ∈ [0;1]: sgn(y) = 1**
* Для любого y принадлежащего отрезку [0;1] функция sgn(y) имеет значение 1 **ЛОЖНО, так как sgn(0)=0**
* отрицание **∃y ∈ [0;1] → sgn(y) != 1** существуют такие y из отрезка [0;1] что функция sgn(y) не равна 1 **ИСТИНА**

**2. ∀n ∈ N>2: ∃x,y,z ∈ N: x^n=y^n + z^n**  
* Для любого натурального n>2 существуют натуральные x,y,z что выполняется равенство x^n=y^n + z^n **сложно оценить на истиность**
* отрицание **∃n ∈ N>2: ∀x,y,z ∈ N→ $x^n != y^n + z^n$** существуют натуральные n>2 что при любых натуральных x,y,z не выполняется равенство $x^n=y^n + z^n$

**3. ∀x ∈ R ∃ X ∈ R: X>x**
* Для любого рационального числа существует подмножество рациональных чисел, такое что подмножество больше чем число **ЛОЖНО, так как может быть X=x, если X - множество из 1 элемента x** 
* отрицание **∃x ∈ R∀X ∈ R→ X<=x** существуют рациональное число из любого подмножества рациональных чисел, такое что подмножество меньше или равно числу. **ИСТИНА**

**4. ∀x ∈ C∄y ∈ C: x>y||x<y**
* Для любого комплексного числа не существует другое комплексное число, которые можно было бы сравнить **ЛОЖНО, 5>4 тоже комплексные числа.**
* отрицание **∃x ∈ C∀y ∈ C→ x=y** существуют комплексное число для любого комплекского числа, такие что числа равны. **ИСТИНА**

**5. ∀y ∈ [0;π/2] ∃ε>0: sin(y) < sin(y+ε)**
* Для любого y из отрезка от 0 до π/2 существует такое положительное ε при котором sin(y) < sin(y+ε) **ЛОЖНО, так как sin(π/2) > sin(π/2 + ε)**
* отрицание **∃y ∈ [0;π/2] ∀ε>0: sin(y) >= sin(y+ε)** существует y из отрезка от 0 до π/2, что при любом положительном ε выполняется неравенство sin(y) >= sin(y+ε) **ИСТИНА**

**6. ∀y ∈ [0;π) ∃ε>0: cos(y) > cos(y+ε)**
* Для любого y из интервала от 0 (вкл.) до π существует такое положительное ε при котором cos(y) > cos(y+ε) **ИСТИНА**
* отрицание **∃y ∈ [0;π) ∀ε>0: cos(y) <= cos(y+ε)** существует y из интервала от 0 (вкл.) до π, что при любом положительном ε выполняется неравенство cos(y) <= cos(y+ε) **ЛОЖНО**

**7. ∃x: x ∉ {N,Z,Q,R,C}**
* Существует число которое не принадлежит множествам N,Z,Q,R,C **ИСТИНА** число может быть гиперкомплексным 
* отрицание **∀x: x ∈ {N,Z,Q,R,C}** любое число принадлежит множествам N,Z,Q,R,C **ЛОЖНО**

## Тема “Множество”

### 1. Даны три множества a,b и с. Необходимо выполнить все изученные виды бинарных операций над всеми комбинациями множеств.

```python
#Зададим множества А и В. Не обязательно числовые
a = {1, 2, 3, 4, "четыре", "пять"}
b = {1, 3, 4, 5, 6, "семь", "восемь"}
c = {4, 6, 8, 9, 10, 'сто'}
```

```python
# операция объединение (логическое И)
print(f'Пересечение a & b = {a & b}')
print(f'Пересечение a & c = {a & c}')
print(f'Пересечение b & c = {b & c}')
print(f'Пересечение a & b & c = {a & b & c}')
```

```python
# операция объединение (логическое ИЛИ)
print(f'Объединение a | b = {a | b}')
print(f'Объединение a | c = {a | c}')
print(f'Объединение b | c = {b | c}')
print(f'Объединение a | b | c = {a | b | c}')
```

```python
# операция разность
print(f'Разность a - b = {a - b}')
print(f'Разность a - c = {a - c}')
print(f'Разность b - c = {b - c}')
print(f'Разность b - a = {b - a}')
print(f'Разность c - a = {c - a}')
print(f'Разность c - b = {c - b}')
print(f'Разность a - b - c = {a - b - c}')
print(f'Разность b - a - c = {b - a - c}')
print(f'Разность c - a - c = {c - a - b}')
```

```python
# операция симметрическая разность
print(f'Симметрическая разность a ^ b = {a ^ b}')
print(f'Симметрическая разность a ^ c = {a ^ c}')
print(f'Симметрическая разность b ^ c = {b ^ c}')
print(f'Симметрическая разность a ^ b ^ c = {a ^ b ^ c}')
```

```python
# операция декартово произведение
from itertools import product

print(f'Декартово произведение a X b = ')
print(list(product(a, b))), print(len(a), end=' * '),  print(len(b), end=' = '), print(len(list(product(a, b))))
print(f'\nДекартово произведение a X c = ')
print(list(product(a, c))), print(len(list(product(a, c))))
print(f'\nДекартово произведение b X c = ')
print(list(product(b, c))), print(len(list(product(b, c))))
print(f'\nДекартово произведение a X b X c = ')
print(list(product(a, b, c)))
print(len(a), end=' * '), print(len(b), end=' * '),  print(len(c), end=' = '), print(len(list(product(a, b, c))))
```
